# Time Frame
Time frame for diffrent trading style

📌 1. Position Trading
| Aspect                 | Recommended Time Frames            |
| ---------------------- | ---------------------------------- |
| **Entry Analysis**     | **Daily (1D)** and **Weekly (1W)** |
| **Trend Confirmation** | Weekly / Monthly                   |
| **Holding Period**     | Weeks to Months / Years            |


📌 2. Swing Trading
| Aspect                 | Recommended Time Frames           |
| ---------------------- | --------------------------------- |
| **Entry Analysis**     | **4-Hour (4H)** or **Daily (1D)** |
| **Trend Confirmation** | Daily (1D)                        |
| **Execution/Timing**   | 1-Hour (1H) or 30-Min             |
| **Holding Period**     | 2–10 days or 1–3 weeks            |


📌 3. Day Trading
| Aspect                 | Recommended Time Frames              |
| ---------------------- | ------------------------------------ |
| **Pre-Market Setup**   | Daily (1D) & 1-Hour (1H) charts      |
| **Entry/Exit**         | **5-Minute** or **1-Minute** charts  |
| **Trend Confirmation** | 15-Min or 1-Hour for bias            |
| **Holding Period**     | Few minutes to hours (same day only) |


📌 4. Scalping
| Aspect                 | Recommended Time Frames              |
| ---------------------- | ------------------------------------ |
| **Entry/Exit**         | **1-Minute (1M)** or **Tick charts** |
| **Trend Confirmation** | 5-Min or 15-Min                      |
| **Holding Period**     | Seconds to few minutes               |


📌 5. Algo Trading
| Aspect                  | Recommended Time Frames                                                           |
| ----------------------- | --------------------------------------------------------------------------------- |
| **Depends on strategy** | 1-Min to Daily                                                                    |
| **Backtesting**         | Use **same timeframe** as the strategy                                            |
| **Examples**            | - Scalping bot: 1M or Tick data<br>- Swing algo: 1H or 4H<br>- Positional bot: 1D |
