Types of charts

1. Line Chart
- Description: Plots only the closing prices over time.
- Best For: Getting a quick view of price direction.
- Pros: Simple and easy to understand.
- Cons: Omits highs, lows, and opens.

✅ Use when you want a quick, clutter-free view of a stock’s trend.


2. Bar Chart (OHLC Chart)
- OHLC: Open, High, Low, Close
- Description: Each bar shows the stock’s open, high, low, and close for a time period.
- Best For: Detailed view of price action and volatility.
- Pros: Gives more insight than a line chart.
- Cons: Can be visually complex for beginners.

3. Candlestick Chart
- Description: Similar to bar charts, but with more visual appeal. Each candle shows open, high, low, and close.
- Green/White Candle: Close > Open (bullish)
- Red/Black Candle: Close < Open (bearish)
- Best For: Spotting patterns like Doji, Hammer, Engulfing, etc.
- Pros: Easier to interpret emotions and reversals.
- Cons: Can become cluttered with many candles.

✅ Most popular chart type among technical analysts and traders.

4. Heikin-Ashi Chart
- Description: A modified candlestick chart that averages price data to smooth out noise.
- Best For: Identifying clear trends and reducing false signals.
- Pros: Makes trends easier to spot.
- Cons: Not suitable for precise entry/exit points (since it uses averaged data).

5. Renko Chart
- Description: Focuses on price movement only; time is not a factor. New "bricks" are added when price moves by a set amount.
- Best For: Filtering out noise and spotting trends.
- Pros: Great for identifying strong trends.
- Cons: Doesn’t show time or volume.

6. Point and Figure (P&F) Chart
- Description: Uses X’s and O’s to represent rising and falling prices, respectively. Ignores time and volume.
- Best For: Identifying support/resistance and breakout patterns.
- Pros: Excellent for long-term trend analysis.
- Cons: Not commonly used by retail traders; more niche.

7. Area Chart
- Description: Similar to a line chart, but the area below the line is filled in with color.
- Best For: Visualizing cumulative price movement.
- Pros: Visually appealing and simple.
- Cons: Like line charts, it lacks detailed OHLC data.

Summary
| Chart Type     | Shows OHLC?   | Time-Based | Best Use                            |
| -------------- | ------------- | ---------- | ----------------------------------- |
| Line           | No            | Yes        | Quick overview of price trends      |
| Bar            | Yes           | Yes        | Detailed price movement             |
| Candlestick    | Yes           | Yes        | Pattern recognition, trend analysis |
| Heikin-Ashi    | No (Averaged) | Yes        | Trend smoothing                     |
| Renko          | No            | No         | Clear trend analysis                |
| Point & Figure | No            | No         | Long-term patterns, breakouts       |
| Area           | No            | Yes        | Visual trend summary                |
