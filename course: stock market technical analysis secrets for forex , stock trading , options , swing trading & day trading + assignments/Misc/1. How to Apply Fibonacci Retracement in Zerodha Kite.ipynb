To apply **Fibonacci Retracement** on **Zerodha Kite charts**, follow these simple steps:

---

## 🪜 How to Apply Fibonacci Retracement in Zerodha Kite

### ✅ On **Desktop/Web Version**:

1. **Login** to [https://kite.zerodha.com](https://kite.zerodha.com)
   and open the chart for the stock/index you want to analyze.

2. **Click “Chart”** for the desired instrument to open the full view.

3. On the **chart toolbar (left or top panel)**, look for the **drawing tools icon** — usually a **pencil or crosshair icon**.

4. Click on it to open the list of **drawing tools**.

5. Scroll or search for **“Fibonacci Retracement”**.

6. **Select it**, and then:

   * **Click once** on the swing **low** (start point).
   * **Click again** on the swing **high** (end point), or vice versa depending on trend.

7. The **Fibonacci levels** (0.0, 0.236, 0.382, 0.5, 0.618, 0.786, 1.0) will now be drawn on the chart.

8. You can **move** or **resize** the levels by dragging the anchors.

---

### 📱 On **Kite Mobile App**:

Unfortunately, the **mobile app has limited charting tools**, and **Fibonacci Retracement is not supported directly** in the current mobile version.

To access advanced tools like Fibonacci, use the **Kite web version on desktop or in a browser on mobile (Desktop mode)**.

---

## 🛠️ Optional: Customize Fibonacci Levels

* Right-click (or double click) on the drawn Fibonacci line.
* A menu will open where you can **customize levels**, colors, and visibility.
* You can enable/disable individual levels as per your strategy.
