In **Zerodha Kite Web or Kite Charts**, if you want to **change the scale of the chart to logarithmic**, here's how to do it:

---

## ✅ How to Change Scale to Logarithmic in Kite Charts

1. **Open Kite Chart**:

   * Go to [https://kite.zerodha.com](https://kite.zerodha.com).
   * Select a stock and open its full chart.

2. **Click on Settings ⚙️ (Chart Properties)**:

   * On the chart page, find the **gear icon (⚙️)** usually located on the **top right corner** of the chart area.

3. **Navigate to the “Scales” tab**:

   * Click **“Scales”** (or sometimes under "Chart Settings").

4. **Enable Log Scale**:

   * You’ll see an option like **“Log Scale”** or **“Logarithmic scale”**.
   * **Check the box** to enable it.

5. **Done!**

   * The y-axis will now change from **linear** to **logarithmic**.

---

## 📌 Notes

* Log scale is useful when you're analyzing stocks that have grown significantly over time — it helps in **visualizing relative % changes**.
* This setting is **per chart**, so you may need to set it again for other charts.

---

If you're using **Kite mobile app**, the option might be limited — log scale is best accessed through the **desktop version**.
