Technical Analysis

Technical analysis in stock trading is a method used to evaluate and predict the future price movements of a stock (or any financial asset) based on its historical price and trading volume data.

It involves analyzing 
- charts and using various statistical tools 
- indicators to identify patterns, trends, and signals that may suggest future activity

Key Concepts in Technical Analysis

1. Price Charts:
- The most basic tool. Includes line charts, bar charts, and candlestick charts.
- Traders use these to spot patterns like head and shoulders, double tops/bottoms, or triangles.

2. Trends:

- Uptrend: Higher highs and higher lows.
- Downtrend: Lower highs and lower lows.
- Sideways trend: Price moves within a horizontal range.

3. Support and Resistance:
- Support: A price level where buying interest is strong enough to stop a price from falling further.
- Resistance: A level where selling interest prevents the price from rising higher.

4. Indicators and Oscillators:
- Tools used to analyze momentum, volatility, and potential reversal points.
- Common ones include:
    - Moving Averages (MA)
    - Relative Strength Index (RSI)
    - MACD (Moving Average Convergence Divergence)
    - Bollinger Bands
    - Volume indicators
5. Volume Analysis:
- High trading volume can confirm trends or reversals.
- Volume spikes often signal the beginning or end of strong price movements.

6. Patterns:
- Continuation patterns (e.g., flags, pennants) suggest a trend will continue.
- Reversal patterns (e.g., double top/bottom, head and shoulders) indicate a potential change in trend.



# Assumptions of Technical Analysis:
- Market discounts everything: All known information is already reflected in the stock’s price.
- Prices move in trends: Once a trend is established, it is likely to continue until something shifts it.
- History repeats itself: Price patterns tend to repeat over time due to investor psychology.
