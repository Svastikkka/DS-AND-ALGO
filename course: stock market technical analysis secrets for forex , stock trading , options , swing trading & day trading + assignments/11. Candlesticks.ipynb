Diffrence between reversal canddle and momentum canddle

🔄 Reversal Candle vs 🚀 Momentum Candle
| Feature                 | **Reversal Candle**                                       | **Momentum Candle**                                    |
| ----------------------- | --------------------------------------------------------- | ------------------------------------------------------ |
| 🧭 **Purpose**          | Indicates a **potential trend change**                    | Shows **continuation** or **strengthening** of a trend |
| ⬅️ **Direction**        | Changes the direction (uptrend → downtrend or vice versa) | Follows the current trend (uptrend or downtrend)       |
| 📈 **Typical Location** | Appears **after an uptrend/downtrend** (at tops/bottoms)  | Appears **during a trend** (in the middle of a move)   |
| 🔥 **Candle Shape**     | Often has **wicks** and **small bodies**                  | Usually has **large bodies** with **small wicks**      |
| 📉 **Examples**         | - Hammer                                                  |                                                        |


1. 🔄 Reversal Candle Example:

    -   fter a 5-day fall, a Hammer appears with a long lower wick → hints at a bullish reversal.
    -   After a rally, a Shooting Star with a long upper wick appears → may signal a bearish reversal.

2. 🚀 Momentum Candle Example:

    -   During a strong uptrend, a big green Marubozu (no wicks) appears → signals strong bullish momentum.
    -   During a breakdown, a big red candle breaks below support on high volume → confirms bearish momentum.

## 🔄 Reversal Candlestick Patterns

1. 📉 Bearish Reversal Patterns (after an uptrend):

| Pattern               | Description                                                                    |
| --------------------- | ------------------------------------------------------------------------------ |
| **Shooting Star**     | Small body at the bottom with a long upper wick                                |
| **Bearish Engulfing** | Large red candle fully engulfs previous green candle                           |
| **Evening Star**      | 3-candle pattern showing uptrend losing steam                                  |
| **Dark Cloud Cover**  | Red candle opens above prior green but closes below its midpoint               |
| **Hanging Man**       | Small body at top with long lower wick (looks like a hammer but after uptrend) |

2. 📈 Bullish Reversal Patterns (after a downtrend):

| Pattern               | Description                                                            |
| --------------------- | ---------------------------------------------------------------------- |
| **Hammer**            | Small body at the top with a long lower wick                           |
| **Bullish Engulfing** | Large green candle engulfs the previous red candle                     |
| **Morning Star**      | 3-candle pattern indicating trend shift from down to up                |
| **Piercing Line**     | Green candle opens below and closes above midpoint of prior red candle |
| **Inverted Hammer**   | Small body at bottom with long upper wick                              |

## 🚀 Momentum Candlestick Patterns

| Pattern                  | Description                                                                |
| ------------------------ | -------------------------------------------------------------------------- |
| **Marubozu**             | A full-bodied candle with little to no wicks (shows strong momentum)       |
| **Three White Soldiers** | 3 consecutive strong green candles with higher closes (bullish momentum)   |
| **Three Black Crows**    | 3 consecutive red candles with lower closes (bearish momentum)             |
| **Bullish Belt Hold**    | A long green candle that opens at the low and closes high                  |
| **Bearish Belt Hold**    | A long red candle that opens at the high and closes low                    |
| **Breakout Candle**      | A large candle that breaks a key support/resistance level with high volume |
