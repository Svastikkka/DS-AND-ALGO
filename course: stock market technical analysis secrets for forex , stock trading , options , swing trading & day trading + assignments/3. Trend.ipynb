# Trend

A trend is the general direction in which a stock or market is moving over a certain period. Trends are one of the core concepts in technical analysis and help traders make decisions on when to enter or exit a trade.

## Types of Trends
1. Uptrend (Bullish Trend):

- The stock price makes higher highs and higher lows.
- Indicates buying pressure and optimism.
- 📈 Example: ₹100 → ₹110 → ₹105 → ₹120.

2. Downtrend (Bearish Trend):

- The price makes lower highs and lower lows.
- Indicates selling pressure and pessimism.
- 📉 Example: ₹120 → ₹110 → ₹115 → ₹100.

3. Sideways Trend (Consolidation):

- The price moves within a horizontal range.
- Neither buyers nor sellers dominate.
- ↔️ Example: ₹100 → ₹102 → ₹98 → ₹101.

## Trend Within a Trend (Fractal Trends)
A trend within a trend refers to smaller trends that exist inside a larger trend. These are often seen when analyzing charts at different timeframes.

Think of it like:
- Primary trend (long-term): 6 months to years.
- Secondary trend (medium-term): Weeks to months.
- Minor trend (short-term): Days to weeks.
