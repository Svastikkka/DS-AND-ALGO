📊 What Is Support and Resistance in Stock Trading?

Support and Resistance are key concepts in technical analysis used to identify price levels where stocks tend to pause, reverse, or break out.

🟩 Support:
- A price level where a stock tends to stop falling and may bounce back up due to buying interest.
- Think of it as a floor for the stock.
- Buyers enter here, believing the stock is "cheap."

Example:
If a stock keeps bouncing back up around ₹200, then ₹200 is its support level.

🟥 Resistance:
- A price level where a stock tends to stop rising and may reverse down due to selling pressure.
- Think of it as a ceiling for the stock.
- Sellers come in, thinking the stock is "expensive."

Example:
If a stock keeps falling back from ₹300, then ₹300 is a resistance level.

🔁 Key Differences:
| Feature            | Support                         | Resistance                     |
| ------------------ | ------------------------------- | ------------------------------ |
| 📉 Price Direction | Stops falling                   | Stops rising                   |
| 💰 Trader Behavior | Buying interest (demand zone)   | Selling interest (supply zone) |
| 📈 Price Reaction  | Bounce upward                   | Bounce downward                |
| 🔁 Role Reversal   | Can become resistance if broken | Can become support if broken   |
| 📌 Example         | ₹200 in a downtrend             | ₹300 in an uptrend             |


🔄 Role Reversal Example:

A stock breaks above resistance ₹300, now that level often becomes new support.

Similarly, if a stock breaks below support ₹200, it may become new resistance.
