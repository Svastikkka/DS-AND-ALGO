Given an integer array nums, move all 0's to the end of it while maintaining the relative order of the non-zero elements.

Note that you must do this in-place without making a copy of the array.

Example 1:
- Input: nums = [0,1,0,3,12]
- Output: [1,3,12,0,0]

Example 2:
- Input: nums = [0]
- Output: [0]

Approach
1. Use a pointer last_non_zero to track the position where the next non-zero element should be placed.

2. Iterate through nums:
    - If the current element is non-zero, place it at nums[last_non_zero] and increment last_non_zero.
3. After placing all non-zero elements at the beginning, fill the rest with 0s.

```python
arr = [0,1,0,3,12]
def moveZeros(arr):
    last_non_zero = 0
    for i in range(len(arr)):
        if arr[i]!=0:
            arr[last_non_zero]=arr[i]
            last_non_zero+=1
    for i in range(last_non_zero,len(arr)):
        arr[i]=0
    return arr
print(moveZeros(arr))
```
