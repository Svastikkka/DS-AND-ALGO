Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

You can return the answer in any order.

Example 1:
- Input: nums = [2,7,11,15], target = 9
- Output: [0,1]
- Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].

Example 2:
- Input: nums = [3,2,4], target = 6
- Output: [1,2]

Example 3:
- Input: nums = [3,3], target = 6
- Output: [0,1]

Approach
1. Initialize an empty dictionary to store each number and its index.
2. Iterate over the array:
    - For each number, calculate its complement (target - num).
    - If the complement is already in the dictionary, you've found the pair — return their indices.
    - Otherwise, store the current number and its index in the dictionary.

3. This ensures that you only loop once through the array (O(n)) and can check for complements in constant time (O(1)).

```python
nums = [2,7,11,15]
target = 9
values = {}
for i, value in enumerate(nums):
    complement = target-value
    if complement in values:
        print([values[complement],i])
        break
    values[value] = i
    
```
