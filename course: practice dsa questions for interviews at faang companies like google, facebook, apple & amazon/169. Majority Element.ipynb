Given an array nums of size n, return the majority element.

The majority element is the element that appears more than ⌊n / 2⌋ times. You may assume that the majority element always exists in the array.

Example 1:
- Input: nums = [3,2,3]
- Output: 3

Example 2:
- Input: nums = [2,2,1,1,1,2,2]
- Output: 2


Approach:

Follow Boyer-Moore Voting Algorithm. Think of this as a vote count between two competing candidates:
- We maintain a candidate and a counter.
- If we find the same candidate, we increment the counter.
- If we find a different number, we decrement the counter.
- When the counter drops to zero, we choose a new candidate.

Because the majority element appears more than half the time, it survives all the "cancellations" caused by other numbers.

```python
nums = [2,2,1,1,1,2,2]
def majorityElement(nums):
    count = 0
    candidate = None
    for i in nums:
        if count==0:
            candidate=i
        if i == candidate:
            count+=1
        else:
            count-=1
    return candidate
print(majorityElement(nums))
```

```python
nums = [2,2,1,1,1,2,2]
def majorityElement(nums):
    distinct_nums = set(nums)
    count=0
    candidate = None

    for i in distinct_nums:
        temp = nums.count(i)
        if temp > count:
            count=temp
            candidate=i
            temp=0
    print(candidate)
majorityElement(nums)
```
