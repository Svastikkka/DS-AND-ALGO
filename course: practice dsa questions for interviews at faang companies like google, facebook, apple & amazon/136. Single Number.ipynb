Given a non-empty array of integers nums, every element appears twice except for one. Find that single one.

You must implement a solution with a linear runtime complexity and use only constant extra space.

Example 1:
- Input: nums = [2,2,1]
- Output: 1

Example 2:

- Input: nums = [4,1,2,1,2]
- Output: 4

Example 3:

- Input: nums = [1]
- Output: 1

Approach:
- XOR-ing all elements will cancel out duplicates, leaving only the unique element.
- Input: nums = [4,1,2,1,2]
```bash
Step-by-step XOR:
0 ^ 4 = 4
4 ^ 1 = 5
5 ^ 2 = 7
7 ^ 1 = 6
6 ^ 2 = 4  ← Result is 4
```


````python
result = 0 
nums = [4,1,2,1,2]
for i in nums:
    result = result ^ i
print(result)
````
