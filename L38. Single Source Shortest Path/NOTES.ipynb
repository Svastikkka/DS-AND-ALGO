# Single Source Shortest Path

### What is Single Source Shortest Path?
- Single Source Shortest Path means: Finding the shortest path from one given source vertex to every other vertex in a graph.
- Example:
    - Imagine a graph:
        ```
        A ---1--- B
        |         |
        4         2
        |         |
        C ---1--- D
        ```
    - If the source is A, SSSP finds:
        | Destination | Shortest Distance |
        | ----------- | ----------------- |
        | A           | 0                 |
        | B           | 1                 |
        | C           | 4                 |
        | D           | 3 (A → B → D)     |

### Why is SSSP important?
- SSSP is used everywhere in real life systems.
- Examples
    - Google Maps / GPS
    - Computer Networks
        -   Routers use SSSP to decide:
        - Which route sends packets fastest
        - Example: OSPF protocol uses Dijkstra
    - Operating Systems
        - Process Scheduling
