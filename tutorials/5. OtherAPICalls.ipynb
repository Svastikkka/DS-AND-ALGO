```python
import logging
from kiteconnect import KiteConnect
import pandas as pd
import datetime as dt
import pandas as pd
```

```python
request_token=open("request_token.txt","r").read()
key_secrets=open("api_key.txt","r").read().split()
kite = KiteConnect(api_key=key_secrets[0])
data = kite.generate_session(request_token=request_token,api_secret=key_secrets[1])
```

```python
# Get dump of all NSE instruments
instruments_dump=kite.instruments("NSE")
instruments_df=pd.DataFrame(instruments_dump)
instruments_df.to_csv("NSE_Instruments.csv",index=False)
```

```python
# Fetch quote details
quote = kite.quote("NSE:INFY")
print(quote)

# Fetch last trading price of an instrument
ltp = kite.ltp("NSE:INFY")
print(ltp)

# Fetch order details
orders = kite.orders()
print(orders)

# Fetch position details
positions = kite.positions()
print(positions)

# Fetch holding details
holdings = kite.holdings()
print(holdings)

'''
Output:

{'NSE:INFY': {'instrument_token': 408065, 'timestamp': datetime.datetime(2025, 4, 4, 21, 19, 40), 'last_trade_time': datetime.datetime(2025, 4, 4, 15, 58, 21), 'last_price': 1451.65, 'last_quantity': 0, 'buy_quantity': 0, 'sell_quantity': 0, 'volume': 0, 'average_price': 0, 'oi': 0, 'oi_day_high': 0, 'oi_day_low': 0, 'net_change': 0, 'lower_circuit_limit': 1306.5, 'upper_circuit_limit': 1596.8, 'ohlc': {'open': 1460.5, 'high': 1472, 'low': 1448, 'close': 1496.5}, 'depth': {'buy': [{'price': 0, 'quantity': 0, 'orders': 0}, {'price': 0, 'quantity': 0, 'orders': 0}, {'price': 0, 'quantity': 0, 'orders': 0}, {'price': 0, 'quantity': 0, 'orders': 0}, {'price': 0, 'quantity': 0, 'orders': 0}], 'sell': [{'price': 0, 'quantity': 0, 'orders': 0}, {'price': 0, 'quantity': 0, 'orders': 0}, {'price': 0, 'quantity': 0, 'orders': 0}, {'price': 0, 'quantity': 0, 'orders': 0}, {'price': 0, 'quantity': 0, 'orders': 0}]}}}
{'NSE:INFY': {'instrument_token': 408065, 'last_price': 1451.65}}
[{'account_id': 'TNM718', 'placed_by': 'TNM718', 'order_id': '250405000008094', 'exchange_order_id': None, 'parent_order_id': None, 'status': 'CANCELLED AMO', 'status_message': None, 'status_message_raw': None, 'order_timestamp': datetime.datetime(2025, 4, 5, 10, 33, 22), 'exchange_update_timestamp': None, 'exchange_timestamp': None, 'variety': 'amo', 'modified': False, 'exchange': 'NSE', 'tradingsymbol': 'INFY', 'instrument_token': 408065, 'order_type': 'LIMIT', 'transaction_type': 'BUY', 'validity': 'DAY', 'validity_ttl': 0, 'product': 'CNC', 'quantity': 1, 'disclosed_quantity': 0, 'price': 1451.65, 'trigger_price': 0, 'average_price': 0, 'filled_quantity': 0, 'pending_quantity': 1, 'cancelled_quantity': 0, 'market_protection': 0, 'meta': {}, 'tag': 'quick', 'tags': ['quick'], 'guid': '01XglK1GsitunWB'}]
{'net': [], 'day': []}
[]
'''
```
