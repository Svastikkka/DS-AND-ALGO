Given two sorted arrays `nums1` and `nums2` of size m and n respectively, return the median of the two sorted arrays.

The overall run time complexity should be `O(log (m+n))`.

Example 1:

- Input: nums1 = [1,3], nums2 = [2]
- Output: 2.00000
- Explanation: merged array = [1,2,3] and median is 2.

Example 2:

- Input: nums1 = [1,2], nums2 = [3,4]
- Output: 2.50000
- Explanation: merged array = [1,2,3,4] and median is (2 + 3) / 2 = 2.5.

```python
nums1 = [1,3]
nums2 = [2]

"""
if odd
median position = (n+1)//2

if even
median positions = n//2 and (n//2)+1

step1: find len of arrs n1 and n2
step2: swap nums1 with nums2 if len(nums1)>len(nums2)
step3: get low and high of nums1; low=0 & high=len(nums1)
step4: iterate till low<=high
step5: inside loop find mid of both arrs
step6: create a partitions
    - l1=nums1[m1-1] only if m1 != 0 
    - r1=nums1[m1] only if m1 != n1
    - l2=nums2[m2-1] only if m2 != 0 
    - r2=nums2[m2] only if m2 != n2
step7:
    check if partition is correct
    - l1<=r2
    - l2<=r1

    if correct:
        - if odd length:
            median=max(l1,l2)

        - if even length:
            median=(max(l1,l2)+min(r1,r2))/2
    if l1>r2:
        move towards left side
        high=m1-1
    else:
        move towards right side
        low=m1+1
"""
if (len(nums1)>len(nums2)):
    nums1,nums2=nums2,nums1
n1=len(nums1)
n2=len(nums2)

low=0
high=len(nums1)

left = (n1 + n2 + 1) // 2

while (low<=high):
    m1=(low + high) // 2
    m2=left - m1

    # partitions
    # left partition values
    l1 = float('-inf') if m1 == 0 else nums1[m1 - 1]
    l2 = float('-inf') if m2 == 0 else nums2[m2 - 1]

    # right partition values
    r1 = float('inf') if m1 == n1 else nums1[m1]
    r2 = float('inf') if m2 == n2 else nums2[m2]

    if (l1<=r2 and l2<=r1):
        # if odd
        if (n1+n2)%2!=0:
            # left side has one extra element
            # biggest element on left = middle number
            print(max(l1,l2))
            break
        else:
            print((max(l1,l2)+min(r1,r2))/2)
            break
    elif (l1>r2):
        high=m1-1
    else:
        low=m1+1
```
