Given an array of integers nums sorted in non-decreasing order, find the starting and ending position of a given target value.

If target is not found in the array, return [-1, -1].

You must write an algorithm with O(log n) runtime complexity.

Example 1:
- Input: nums = [5,7,7,8,8,10], target = 8
- Output: [3,4]

Example 2:
- Input: nums = [5,7,7,8,8,10], target = 6
- Output: [-1,-1]

Example 3:
- Input: nums = [], target = 0
- Output: [-1,-1]
 



Approach:

Note: This is a binary search problem because we need to find positions in a sorted array in O(log n) time.

Think binary search when arrary is sorted 

```python
nums = [5,7,7,8,8,8,10]
target = 8
indexes=[]
for i in range(len(nums)):
    if nums[i]  == target:
        indexes.append(i)
print(indexes[0],indexes[len(indexes)-1])
```

```python
def searchRange(nums, target):
    def findLeft():
        left, right = 0, len(nums) - 1
        pos = -1
        while left <= right:
            mid = (left + right) // 2
            if nums[mid] >= target:
                right = mid - 1
            else:
                left = mid + 1
            if nums[mid] == target:
                pos = mid
        return pos

    def findRight():
        left, right = 0, len(nums) - 1
        pos = -1
        while left <= right:
            mid = (left + right) // 2
            if nums[mid] <= target:
                left = mid + 1
            else:
                right = mid - 1
            if nums[mid] == target:
                pos = mid
        return pos

    return [findLeft(), findRight()]
print(searchRange([5,7,7,8,8,10], 8))  # Output: [3, 4]
```

```python
def binarysearch(arr,target):
    def leftmost():
        left = 0
        right = len(arr)-1
        pos = -1
        while left<=right:
            mid = (left+right)//2
            if arr[mid] < target:
                left = mid + 1
            else:
                right = mid - 1
            if arr[mid] == target:
                pos = mid
        return pos             

    def rightmost():
        left = 0
        right = len(arr)-1
        pos = -1
        while left<=right:
            mid = (left+right)//2
            if arr[mid] > target:
                right = mid - 1
            else:
                left = mid + 1
            if arr[mid] == target:
                pos = mid
        return pos
    return [leftmost(), rightmost()]
print(binarysearch([5,7,7,8,8,10], 8))
```
