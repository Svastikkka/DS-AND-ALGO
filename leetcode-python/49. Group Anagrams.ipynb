Given an array of strings strs, group the anagrams together. You can return the answer in any order.

Example 1:
- Input: strs = ["eat","tea","tan","ate","nat","bat"]
- Output: [["bat"],["nat","tan"],["ate","eat","tea"]]
- Explanation: There is no string in strs that can be rearranged to form "bat". The strings "nat" and "tan" are anagrams as they can be rearranged to form each other. The strings "ate", "eat", and "tea" are anagrams as they can be rearranged to form each other.

Example 2:
- Input: strs = [""]
- Output: [[""]]

Example 3:
- Input: strs = ["a"]
- Output: [["a"]]

Approach:
My Approach
- Use dictionary
- Sort each word and in time of iterating if sorted word match with key then put it inside the list


```python
strs = ["eat","tea","tan","ate","nat","bat"]
anagrams = {}
for i in strs:
    if "".join(sorted(i)) not in anagrams:
        anagrams["".join(sorted(i))] = []
    anagrams["".join(sorted(i))].append(i)

print(anagrams)
result = []
for i in anagrams.values():
    result.append(i)
print(result)
```

```python
def groupAnagrams(strs):
    anagrams = {}  # regular dictionary

    for word in strs:
        key = ''.join(sorted(word))  # create anagram signature

        if key in anagrams:
            anagrams[key].append(word)
        else:
            anagrams[key] = [word]

    return list(anagrams.values())
strs = ["eat", "tea", "tan", "ate", "nat", "bat"]
print(groupAnagrams(strs))
```
