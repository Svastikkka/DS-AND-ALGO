Given two binary strings a and b, return their sum as a binary string.

Example 1:
- Input: a = "11", b = "1"
- Output: "100"

Example 2:
- Input: a = "1010", b = "1011"
- Output: "10101"

Approach:

Method 1
- Use inbuilt python bin function

Methos 2
- Create your own decimal to binary and binary to decimal function

```python
def addBinary(a,b):
    result = bin(int(a,2)+int(b,2))[2:]
    return result
addBinary("1010","1011")
```
