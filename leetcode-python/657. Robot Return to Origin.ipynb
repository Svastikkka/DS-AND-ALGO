```python
position = [0,0]
moves = "UD"
for i in moves:
    if i ==  "L":
        position[0]=position[0]-1
    if i == "R":
        position[0]=position[0]+1
    if i == "U":
        position[1]=position[1]+1
    if i == "D":
        position[1]=position[1]-1
print(position)
```

```python
from collections import Counter

def judgeCircle(moves):
    c = Counter(moves)
    return c['L'] == c['R'] and c['U'] == c['D']
print(judgeCircle("UDL"))
```
